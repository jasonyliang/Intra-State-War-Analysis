```python
# We only want to look at intra wars
intra_data = data[data['type_of_conflict'] == 3]
intra_data.head()
```

```python
# lethality over the years
plt.scatter(intra_data['year'], intra_data['bd_best'])
```

```python
# zoom in
# remove outliers
lower_leth = intra_data[intra_data['bd_best']<10000]
plt.scatter(lower_leth['year'], lower_leth['bd_best'])
```

```python
# regions
def pie_chart_region(wars, startyears = None, endyears=None):
    if endyears and startyears:
        wars = wars[(wars["year"] < endyears) &(wars["year"] > startyears)]
    elif endyears:
        wars = wars[wars["year"] < endyears]
    elif startyears:
        wars = wars[wars["year"] > startyears]
    regions = {1:"Europe", 2: "Middle East", 3: "Asia", 4: "Africa", 5: "Americas"}
    size = []
    names = []
    for num, region in regions.items():
        numWars = wars[wars['region']==str(num)].count(axis=0)[0]
        names.append(region)
        size.append(numWars)
    plt.pie(size, labels=names, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    plt.tight_layout()
```

```python
# post 1990
pie_chart_region(intra_data, startyears=1990)
```

```python
# all intra-war in the data set
pie_chart_region(intra_data)
```
