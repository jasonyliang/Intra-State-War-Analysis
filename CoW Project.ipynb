```python
# libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
# import data
data = pd.read_csv("data/Intra-StateWarData_v4.1.csv", encoding='latin-1')
data.head()
```

### Central Question of Concern:
How has the lethality of intra-state violence (civil wars and internal conflicts) changed over time?

## Difference between regions on the number of intra-wars 

```python
# regions
def pie_chart_region(wars, startyears = None, endyears=None):
    if endyears and startyears:
        wars = wars[(wars["StartYear1"] < endyears) &(wars["StartYear1"] > startyears)]
    elif endyears:
        wars = wars[wars["StartYear1"] < endyears]
    elif startyears:
        wars = wars[wars["StartYear1"] > startyears]
    regions = {1:"W. Hemisphere", 2: "Europe", 4: "Africa", 6: "Middle East", 7: "Asia", 9:"Oceania"}
    size = []
    names = []
    for num, region in regions.items():
        numWars = wars[wars['WhereFought']==num].count(axis=0)[0]
        names.append(region)
        size.append(numWars)
    plt.pie(size, labels=names, autopct='%1.1f%%', shadow=True, startangle=140)
    plt.axis('equal')
    plt.tight_layout()
```

```python
pie_chart_region(data)
```

```python
# pre-1990
pie_chart_region(data, endyears=1990)
```

```python
#post-1990 to 2007
pie_chart_region(data, startyears=1990)
```

## Lethality

```python
def convert_lethal(row):
    sideA, sideB = int(row['SideADeaths'].replace(',','')), int(row['SideBDeaths'].replace(',',''))
    sideA = None if sideA == -8 or sideA == -9 else sideA
    sideB = None if sideB == -8 or sideB == -9 else sideB
    if sideA and sideB:
        return int(sideA+sideB)
    elif sideA:
        return int(sideA)
    elif sideB:
        return int(sideB)
    else:
        return None

data['totLethality'] = data.apply(lambda row: convert_lethal(row), axis = 1)
```

```python
data['totLethality'].describe()
```

```python
leth_data = data[data['totLethality']!= None]
plt.scatter(leth_data['StartYear1'], leth_data['totLethality'])
plt.title("Lethality of war over time (Based on the start of each war)")
plt.ylabel("Lethality")
plt.xlabel("Start Year")
```

```python
x = leth_data['StartYear1'].unique()
y = [leth_data['totLethality'][leth_data.StartYear1 == i].mean() for i in x]
plt.plot(x, y)
plt.title("Average lethality of wars given their start year")
plt.ylabel("Lethality")
plt.xlabel("Start Year")
```

```python
def avg_decade(data, var, novar, start=1810, end=2010):
    decades = [(i, i+10) for i in range(start, end, 10)]
    avg = []
    decade = []
    for t, t_1 in decades:
        wars = data[(data.StartYear1 >= t) & (data.StartYear1 < t_1) & (data[var]!= novar)]
        v = []
        for w in wars[var]:
            if w != None:
                v.append(w)
        
        a_val = np.nanmean(v) if np.nanmean(v) == np.nanmean(v) else 0
        avg.append(a_val)
        decade.append(t)
    return decade, avg
```

```python
decade, avg_leth = avg_decade(leth_data,'totLethality', None)
plt.plot(decade, avg_leth)
```

### Lethality by Region

```python
regions = {1:"W. Hemisphere", 2: "Europe", 4: "Africa", 6: "Middle East", 7: "Asia", 9:"Oceania"}
region_data = []
for key, value in regions.items():
    reg_data = data[data['WhereFought'] == key]
    region_data.append((value, reg_data))
```

```python
plt.figure(figsize=(15,8))
for (reg, dta) in region_data:
    decade, avg_leth = avg_decade(dta, 'totLethality', None)
    plt.plot(decade, avg_leth, label=f"{reg}", ls='-', marker='.', ms=12)
plt.legend()
```

```python
# average lethality from 1970 to present
region_data_70 = []
for key, value in regions.items():
    reg_data_70 = data[(data['WhereFought'] == key) & (data['StartYear1'].astype(float) >= 1970)]
    region_data_70.append((value, reg_data_70))
```

```python
plt.figure(figsize=(15,8))
for (reg, dta) in region_data_70:
    decade, avg_leth = avg_decade(dta, 'totLethality', None, start=1970)
    plt.plot(decade, avg_leth, label=f"{reg}", ls='-', marker='.', ms=12)
plt.legend()
```

### Duration of war

```python
# Duration
def war_duration(row):
    start = row['StartYear1']
    end = row['EndYear1']
    if not(end == end) or int(end) == -7: # still going on as of 2007
        return -7 # we only count concluded wars
    else:
        return end - start

data['Duration'] = data.apply(lambda row: war_duration(row), axis=1)
```

```python
# distribution of durations of intra-state wars
plt.hist(data['Duration'][data.Duration != -7])
```

```python
year, dur = data['StartYear1'][data.Duration!= -7] , data['Duration'][data.Duration!= -7] 
plt.scatter(year, dur)
```

```python
# Average by decades 1810 - 2000
decades = [(i, i+10) for i in range(1810, 2010, 10)]
avg_duration = []
decade = []
for t, t_1 in decades:
    wars = data[(data.StartYear1 >= t) & (data.StartYear1 < t_1) & (data.Duration!= -7)]
    durations = []
    for w in wars['Duration']:
        durations.append(w)
    avg_duration.append(np.mean(durations))
    decade.append(t)
```

```python
decade, avg_duration = avg_decade("Duration", -7)
plt.plot(decade, avg_duration)
```

```python
# duration vs lethality
plt.scatter(data['Duration'][data.Duration!= -7], data['totLethality'][data.Duration!= -7])
plt.title("Duration of war vs lethality")
```
